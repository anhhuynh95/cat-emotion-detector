### 3.2 Augmentation for Diversity
First, I’ll use augmentation to boost `happy` from 98 to 150, adding variety in angles and lighting. I’ll leave `sad` (132) and `angry` (139) as-is for now, since they’re closer to 150. I’ll use flips, rotations, and brightness tweaks to create new `happy` images.

```python
import imgaug.augmenters as iaa
import cv2
import numpy as np
import os
import pandas as pd

# Load the CSV
df = pd.read_csv('master_dataset_labels_v1.csv')

# Augmentation pipeline
seq = iaa.Sequential([
    iaa.Fliplr(0.5),  # 50% horizontal flip
    iaa.Affine(rotate=(-10, 10)),  # Rotate -10 to 10 degrees
    iaa.Multiply((0.8, 1.2))  # Brightness tweak
])

# Create a folder for augmented images
aug_folder = 'augmented_images'
if not os.path.exists(aug_folder):
    os.makedirs(aug_folder)

# Function to augment images
def augment_images(df_subset, target_count, emotion):
    current_count = len(df_subset)
    if current_count >= target_count:
        return df_subset.sample(target_count)  # Downsample if needed
    images = [cv2.imread(f) for f in df_subset['filename']]
    aug_needed = target_count - current_count
    aug_images = seq.augment_images(images * int(np.ceil(aug_needed / current_count)))[:aug_needed]
    new_rows = []
    for i, aug_img in enumerate(aug_images):
        aug_path = os.path.join(aug_folder, f'aug_{emotion}_{i}.jpg')
        cv2.imwrite(aug_path, aug_img)
        new_rows.append({'filename': aug_path, 'emotion': emotion})
    return pd.concat([df_subset, pd.DataFrame(new_rows)])

# Target 150 for happy, keep sad and angry as-is
df_happy = augment_images(df[df['emotion'] == 'happy'], 150, 'happy')
df_sad = df[df['emotion'] == 'sad']
df_angry = df[df['emotion'] == 'angry']

# Combine and save intermediate CSV
df_intermediate = pd.concat([df_happy, df_sad, df_angry])
df_intermediate.to_csv('intermediate_dataset_labels_v1.csv', index=False)

print("Balance after augmentation:")
print(df_intermediate['emotion'].value_counts())
```

### 3.3 Balancing with SMOTE
Now I have 150 happy, 132 sad, and 139 angry so the distance between them is closer, but still not balanced. I’ll use SMOTE to get to 200 per class (600 total). SMOTE creates synthetic samples by interpolating in feature space, so I’ll preprocess the images (resize to 64x64, grayscale, flatten) first, apply SMOTE, then save the new samples as images. This is my first time using SMOTE for images, so let’s see how it goes!

```python
import cv2
import numpy as np
import pandas as pd
import os
from imblearn.over_sampling import SMOTE

# Preprocess function for SMOTE
def preprocess_image(filepath):
    img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    if img is None:
        return np.zeros((64, 64)).flatten()
    img = cv2.resize(img, (64, 64))
    return img.flatten()

# Load intermediate dataset
df_intermediate = pd.read_csv('intermediate_dataset_labels_v1.csv')

# Preprocess images to feature vectors
X = np.array([preprocess_image(row['filename']) for _, row in df_intermediate.iterrows()])
y = df_intermediate['emotion'].values

# Apply SMOTE to balance to 200 per class
smote = SMOTE(sampling_strategy={'happy': 200, 'sad': 200, 'angry': 200}, random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# Create a folder for SMOTE-generated images
smote_folder = 'smote_images'
if not os.path.exists(smote_folder):
    os.makedirs(smote_folder)

# Save SMOTE-generated samples as images
new_rows = []
for i, (features, label) in enumerate(zip(X_smote, y_smote)):
    if i < len(df_intermediate):
        new_rows.append({'filename': df_intermediate['filename'].iloc[i], 'emotion': label})
    else:
        img = features.reshape(64, 64).astype(np.uint8)
        smote_path = os.path.join(smote_folder, f'smote_{label}_{i-len(df_intermediate)}.jpg')
        cv2.imwrite(smote_path, img)
        new_rows.append({'filename': smote_path, 'emotion': label})

# Create final balanced dataset
df_balanced = pd.DataFrame(new_rows)
df_balanced.to_csv('balanced_dataset_labels_v1.csv', index=False)

print("Final balance after SMOTE:")
print(df_balanced['emotion'].value_counts())
```

### 3.6 Fixing the Dataset with Enhanced Augmentation
Since the SMOTE-generated images aren’t usable, I’ll go back to the intermediate dataset (`happy: 150`, `sad: 132`, `angry: 139`) and balance to 200 per class (600 total) using only augmentation. I’ll also add more transformations (cropping and contrast adjustments) to the existing flips, rotations, and brightness tweaks to increase diversity in angles, framing, and lighting.

```python
import imgaug.augmenters as iaa
import cv2
import numpy as np
import os
import pandas as pd

# Load the intermediate dataset (before SMOTE)
df_intermediate = pd.read_csv('intermediate_dataset_labels_v1.csv')

# Enhanced augmentation pipeline with GammaContrast
seq = iaa.Sequential([
    iaa.Fliplr(0.5),  # 50% horizontal flip
    iaa.Affine(rotate=(-10, 10)),  # Rotate -10 to 10 degrees
    iaa.Multiply((0.8, 1.2)),  # Brightness tweak
    iaa.Crop(percent=(0, 0.1)),  # Random crop up to 10% of the image
    iaa.GammaContrast((0.8, 1.2))  # Adjust contrast using gamma correction
])

# Create a folder for augmented images 
aug_folder = 'augmented_images'
if not os.path.exists(aug_folder):
    os.makedirs(aug_folder)

# Function to augment images
def augment_images(df_subset, target_count, emotion):
    current_count = len(df_subset)
    if current_count >= target_count:
        return df_subset.sample(target_count)  # Downsample if needed
    images = [cv2.imread(f) for f in df_subset['filename']]
    aug_needed = target_count - current_count
    aug_images = seq.augment_images(images * int(np.ceil(aug_needed / current_count)))[:aug_needed]
    new_rows = []
    for i, aug_img in enumerate(aug_images):
        aug_path = os.path.join(aug_folder, f'aug_{emotion}_{i+len(df_subset)}.jpg')  # Offset index to avoid overwriting
        cv2.imwrite(aug_path, aug_img)
        new_rows.append({'filename': aug_path, 'emotion': emotion})
    return pd.concat([df_subset, pd.DataFrame(new_rows)])

# Target 200 for all classes
df_happy = augment_images(df_intermediate[df_intermediate['emotion'] == 'happy'], 200, 'happy')
df_sad = augment_images(df_intermediate[df_intermediate['emotion'] == 'sad'], 200, 'sad')
df_angry = augment_images(df_intermediate[df_intermediate['emotion'] == 'angry'], 200, 'angry')

# Combine and save final balanced CSV
df_balanced = pd.concat([df_happy, df_sad, df_angry])
df_balanced.to_csv('balanced_dataset_labels_v2.csv', index=False)

print("Final balance after enhanced augmentation (updated with GammaContrast):")
print(df_balanced['emotion'].value_counts())
```

### 3.8 Preprocessing the Data
Now with 600 balanced images, I’ll preprocess them for SVM training. I’ll extract HOG (Histogram of Oriented Gradients) features to capture edges and shapes (e.g., ear position, eye shape), which are key for detecting cat emotions, as noted in my research. HOG features will reduce the dimensionality (to 648 features per image) and provide more meaningful input for SVM. I’ll also encode the labels (happy, sad, angry) into numbers because SVM expects numerical labels, not strings. Then, I’ll split 80/20 for training.

```python
import cv2
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from skimage.feature import hog

# Load balanced CSV
df_balanced = pd.read_csv('balanced_dataset_labels_v2.csv')

# Preprocess function with HOG feature extraction
def preprocess_image_hog(filepath):
    # Read image in grayscale
    img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    if img is None:
        return np.zeros(648)  # Return zero vector if image fails to load
    # Resize to 64x64 (HOG works better with consistent sizes)
    img = cv2.resize(img, (64, 64))
    # Extract HOG features
    hog_features = hog(img,
                       orientations=9, # calculates gradients in 9 different angle bins (for example, 0°, 20°, 40°, etc.)
                       pixels_per_cell=(8, 8), # each small cell is 8x8 pixels. In each cell, it calculates a histogram of gradients.
                       cells_per_block=(2, 2), # cells are grouped into bigger blocks (2x2 cells = 16x16 pixels). This helps normalize gradient values and make it robust to lighting changes.
                       block_norm='L2-Hys') # the normalization technique to keep values stable.
    return hog_features

# Process images with HOG
X = np.array([preprocess_image_hog(row['filename']) for _, row in df_balanced.iterrows()])
y = df_balanced['emotion'].values

# Encode the labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Print the mapping to understand the encoding
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

# Split with encoded labels
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42)
```

### 4.2 Reflection on Linear Kernel Results
The linear kernel’s accuracy came out around **46.6%**, which is quite low.
It performed slightly better on the **happy** class (22 out of 40 correct), but **angry** and especially **sad** categories showed many misclassifications.

Looking at the confusion matrix:
- Sad cats were frequently misclassified as angry (14 times) or happy (11 times).
- Angry cats were sometimes predicted as sad (13 times).
- This indicates that even though I used HOG features to capture important edges and shapes, the data **cannot be easily separated with a straight (linear) decision boundary**.

Before switching to a non-linear kernel like RBF, I will try to tune the **C parameter** for the linear kernel to see if any improve.
I will try different C values (0.1, 1, 10) to see if this improves linear separability.
If that still doesn’t help, I will proceed to try the **RBF kernel**, which allows for more complex decision boundaries.

### 4.3 Tuning C for Linear Kernel
I will try different values of C: **[0.01, 0.1, 1, 10, 100]** — this range allows me to explore both relaxed boundaries (small C, allowing misclassifications for a smoother margin) and strict boundaries (larger C, focusing on correctly classifying all training samples but risking overfitting).

To avoid manually testing each value, I will use `GridSearchCV`, a tool from scikit-learn that automates this process.

    GridSearchCV systematically tests all parameter combinations and evaluates them using cross-validation (cv=3).
    According to the Scikit-learn documentation, GridSearchCV helps avoid overfitting by validating on multiple folds and ensures that the best hyperparameters are chosen based on stable performance rather than random splits.
    
Ref: https://scikit-learn.org/stable/modules/grid_search.html

```python
from sklearn.model_selection import GridSearchCV

# Parameter tuning for linear kernel
param_grid_linear = {
    'C': [0.01, 0.1, 1, 10, 100],
    'kernel': ['linear']
}

grid_linear = GridSearchCV(SVC(random_state=42), param_grid_linear, refit=True, verbose=2, cv=3, n_jobs=-1)
grid_linear.fit(X_train, y_train)

print("Best Parameters for Linear Kernel:", grid_linear.best_params_)

# Evaluate best linear model
best_linear_svm = grid_linear.best_estimator_
y_pred_best_linear = best_linear_svm.predict(X_test)

print("Tuned Linear Kernel SVM Evaluation:")
print(classification_report(y_test, y_pred_best_linear, target_names=label_encoder.classes_))
print("Best Accuracy (Linear Kernel after tuning):", accuracy_score(y_test, y_pred_best_linear))
```

### 4.4 Reflection on Linear Kernel Tuning
After tuning the C parameter with **GridSearchCV**, the best value was found to be **C = 1**, which is actually the default value I used the first time.
So unfortunately, nothing really improved… The accuracy stayed around 47%, and misclassifications still happened, especially between sad and angry.

=> Therefore, I will now move on to using the **RBF kernel**, which is more flexible and can capture complex, non-linear decision boundaries.
I expect this to help the model learn subtle differences in cat facial expressions and posture more effectively.

### 4.5 SVM Model (RBF Kernel)
Since the linear kernel didn’t perform well and tuning didn’t help, it’s time to move on to the RBF (Radial Basis Function) kernel.
**Why?**
- The RBF kernel is more flexible and can model non-linear decision boundaries.
- It’s often used in image classification tasks where patterns are not linearly separable.

So now I’ll train an SVM with the RBF kernel using default settings first (`C=1.0`, `gamma='scale'`):

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Train SVM with RBF kernel
svm_rbf = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_rbf.fit(X_train, y_train)

# Predict
y_pred_rbf = svm_rbf.predict(X_test)

# Evaluate
print("RBF Kernel SVM Results:")
cm = confusion_matrix(y_test, y_pred_rbf)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel("Predicted Emotion")
plt.ylabel("Actual Emotion")
plt.title("Confusion Matrix (Linear Kernel SVM)")
plt.show()
print(classification_report(y_test, y_pred_rbf, target_names=label_encoder.classes_))
print("Accuracy (RBF Kernel):", accuracy_score(y_test, y_pred_rbf))
```

### 4.6 Reflection on RBF Kernel Results
After switching to the RBF kernel, I immediately noticed improvements compared to the linear kernel.

**Key observations from the results:**
- **Accuracy** increased from around **46% (linear)** to **58% (RBF)**.
- The **happy** class performed the best with a **precision and recall of 0.75**, meaning the model confidently identified happy cats with fewer mistakes.
- The **angry** class improved too, with a **recall of 0.60** — 24 out of 40 angry cats were correctly classified.
- The **sad** category still struggles a bit (**recall = 0.40**), showing that the model sometimes confuses sad with angry or happy, probably because those facial features and postures overlap more.

**Confusion matrix explanation:**
- The model **misclassifies many sad cats as angry (19 times)** — this confirms that sad and angry features are sometimes visually close.
- Happy cats are the most easily identified, possibly because they have more distinctive features (ears forward, soft eyes).

**Conclusion:**
The RBF kernel made a noticeable improvement by capturing non-linear patterns in the HOG feature space. However, there’s still room for improvement — especially in sad and angry cats.

### 4.7 Hyperparameter Tuning for RBF Kernel
While the RBF kernel already improved results, I believe the result probably can be improved better by fine-tuning its hyperparameters which is `C` and `gamma`: 
- C (Regularization parameter): Controls how much the model tries to avoid misclassification.
- Gamma (Kernel coefficient): Controls how far the influence of a single training example reaches.

Ref: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

I’ll use **GridSearchCV** again to test combinations of `C` and `gamma` values and pick the best one.

```python
from sklearn.model_selection import GridSearchCV

param_grid_rbf = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 'scale'],
    'kernel': ['rbf']
}

grid_rbf = GridSearchCV(SVC(random_state=42), param_grid_rbf, refit=True, verbose=2, cv=3, n_jobs=-1)
grid_rbf.fit(X_train, y_train)

print("Best Parameters for RBF Kernel:", grid_rbf.best_params_)

# Evaluate best RBF model
best_rbf_svm = grid_rbf.best_estimator_
y_pred_best_rbf = best_rbf_svm.predict(X_test)

print("Tuned RBF Kernel SVM Evaluation:")
cm_rbf_tuned = confusion_matrix(y_test, y_pred_best_rbf)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_rbf_tuned, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel("Predicted Emotion")
plt.ylabel("Actual Emotion")
plt.title("Confusion Matrix (RBF Kernel SVM Tuned)")
plt.show()
print(classification_report(y_test, y_pred_best_rbf, target_names=label_encoder.classes_))
print("Best Accuracy (RBF Kernel after tuning):", accuracy_score(y_test, y_pred_best_rbf))
```

### 4.8 Reflection on RBF Hyperparameter Tuning
The best parameters found were **C = 10** and **gamma = 0.1**. After tuning, accuracy improved from **58%** to **68%—a really clear improvement!
Looking at the confusion matrix, I can say:
- The **happy** class became even more stable, with **34 out of 40** correct predictions (**recall = 0.85**).
- The **sad** class improved **from 40% recall to 62% recall**, showing that tuning helped reduce confusion with angry cats.
- **Angry** predictions also became more balanced (**precision = 0.66**).

=> This confirms that **hyperparameter tuning really matters**! With RBF and the right parameters, my model now captures the subtle patterns in cat emotions much better.
